# Renomeando colunas

Existem duas formas de renomear colunas de um dataset, a apresentada abaixo que necessita que todas as colunas sejam renomeadas para ter efeito e a segunda forma é renomar as colunas escolhidas, e.g: data.rename({'ColunaAtual':'ColunaQueVaiVirar'})

Similar ao comando "AS" de um banco de dados.

```python
data.columns = ['IdPassageiro', 'Sobreviveu', 'Classe', 'Nome', 'Sexo', 'Idade', 
                'IrmaosConjuge','PaisFilhos', 'Bilhete', 'Tarifa', 'Cabine', 'Embarque']
```

# Groupby

Groupby serve para agrupar as informações, porém é importante ressaltar que se executar somente o comando dataset.groupby() será retornado um código hexadecimal representando o dataset com groupby, por isso é necessario apresentar este através de uma função de chamda e.g: tail(), head(), mean(), max(), min(), count() e etc.

Similar ao comando "Groupby" de um banco de dados.

```python
data[['Sexo','Sobreviveu', 'Classe', 'Idade', 'Tarifa']].groupby('Sexo').mean()
```

```python
data[['Embarque', 'Sobreviveu', 'Classe', 'Idade', 'Tarifa']].groupby('Embarque').mean()
```

```python
data[['Embarque', 'Sobreviveu', 'Classe', 'Idade', 'Tarifa']].groupby('Sobreviveu').mean()
```

```python
data[['Embarque', 'Sobreviveu', 'Classe', 'Idade', 'Tarifa']].groupby('Classe').mean()
```

```python
data[['Embarque', 'Sexo', 'Sobreviveu', 'Classe', 'Idade', 'Tarifa']].groupby(['Embarque', 'Sexo']).mean()
```

Loc() serve para localizar uma ou mais linhas dentro de um dataset através de uma condição.

Similar ao comando "Where" de um banco de dados.

```python

data[['Embarque', 'Sexo', 'Sobreviveu', 'Classe', 'Idade', 'Tarifa']].loc[data['Sexo'] == 'female'].groupby(['Tarifa', 'Sexo']).mean()
```

```python
data[['Embarque', 'Sexo', 'Sobreviveu', 'Classe', 'Idade', 'Tarifa']].loc[data['Sexo'] == 'male']\
.groupby(['Embarque', 'Sexo', 'Sobreviveu', 'Tarifa']).mean()
```

## Crosstab

Crosstab serve para cruzar as informações, este contém três parametros sendo eles: a informação das linhas, a informação das colunas e se contá o total das informações (linha + coluna) e é possível estilizar as tabelas resultantes do crosstab para uma melhor visualização.

Similar ao "crosstab" do postgresql

```python
pd.crosstab(data['Sobreviveu'], data['Classe'], margins=True)
```

```python
pd.crosstab(data['Sobreviveu'], data['Sexo'], margins=True).style.background_gradient(cmap='Blues')
```

```python
pd.crosstab(data['Sobreviveu'], data['Sexo'], margins=True).style.background_gradient(cmap='OrRd')
```

```python
pd.crosstab(data['Sexo'], data['Tarifa'])
```

```python
pd.crosstab(data['Classe'], data['Sobreviveu'])
```

```python
pd.crosstab(data['Sexo'], data['Idade'])
```

```python
pd.crosstab(data['Idade'], data['Embarque'])
```

```python
pd.crosstab(data['Sobreviveu'], data['Bilhete'], margins=True).style.background_gradient(cmap='GnBu')
```

```python
pd.crosstab(data['Sobreviveu'], data['Embarque'], margins=True).style.background_gradient(cmap='PuRd')
```

## Ordenar dados

Sort_values serve para ordenar os dados, podendo ser por uma coluna do dataset ou pelos index gerados no dataset.

Similar ao order by de um banco de dados.

```python
data[['Idade', 'Sexo', 'Classe', 'Sobreviveu']].sort_values('Idade', ascending=True)
```

```python
data[['Idade', 'Sexo', 'Classe', 'Sobreviveu']].sort_index(ascending=True)
```

```python
data[['Idade', 'Sexo', 'Classe', 'Sobreviveu']].sort_index(ascending=False)
```

```python
data[['Nome', 'Sexo', 'Idade', 'Tarifa', 'Cabine']].sort_values('Tarifa', ascending=True)
```

```python
data[['Nome', 'Sexo', 'Idade', 'Tarifa', 'Cabine']].sort_values('Idade', ascending=True)
```
