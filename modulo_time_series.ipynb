## Time delta (diferença de tempo)

```python
antes = datetime.datetime.now()
```

```python
depois = datetime.datetime.now()
```

```python
print(antes,depois)
```

```python
## timedelta
diferenca = depois-antes
```

```python
## extrair informações
diferenca.seconds
```

```python
diferenca.microseconds
```

```python
datetime.timedelta(days=10)
```

```python
datetime.timedelta(seconds=10)
```

## String para data hora

```python
str_data_hora = '01/04/20 17:14:00'
```

```python
convertida_para_datetime = datetime.datetime.strptime(str_data_hora, '%m/%d/%y %H:%M:%S')
```

```python
convertida_para_datetime
```

```python
print(convertida_para_datetime)
```
