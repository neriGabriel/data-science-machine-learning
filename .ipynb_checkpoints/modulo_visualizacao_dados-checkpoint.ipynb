## Preparando os dados

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('titanic/train.csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
data.columns = ['IdPassageiro', 'Sobreviveu', 'Classe', 'Nome', 'Sexo', 'Idade', 
                'IrmaosConjuge','PaisFilhos', 'Bilhete', 'Tarifa', 'Cabine', 'Embarque']
```

```python
data['Sexo'].replace({'male': 'homem', 'female':'mulher'}, inplace=True)
```

```python
data['Cabine'] = data['Cabine'].apply(lambda x: x[-1] if pd.notna(x) else np.nan)
```

```python
data['Cabine']
```

## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
##imagens que executarmos vãos aparecer no notebook
%matplotlib inline
```

```python
##histograma da idade
plt.hist(data['Idade'].dropna())
plt.title('Distribuição das Idades')
plt.ylabel('Pessoas')
plt.xlabel('Idade')
plt.show()
```

```python
plt.hist(data['Classe'].dropna())
plt.title('Distribuição de Classes')
plt.ylabel('Pessoas')
plt.xlabel('Classes')
plt.show()
```

### Subplots - 1ª forma

```python
# Parametros: linha, colunas, plot
# plt.figure(figsize=(6,3), dpi=150)
plt.subplot(1, 2, 1)
plt.hist(data['Idade'].dropna())
plt.title('Distribuição das Idades')

plt.subplot(1, 2, 2)
plt.hist(data['Classe'].dropna())
plt.title('Distribuição das Classes')

plt.show()
```

```python
# Parametros: linha, colunas, plot
# plt.figure(figsize=(6,3), dpi=150)
plt.subplot(2, 1, 1)
plt.hist(data['Idade'].dropna())
plt.title('Distribuição das Idades')

plt.subplot(2, 1, 2)
plt.hist(data['Classe'].dropna())
plt.title('Distribuição das Classes')

# tira a distancia dos gráficos para não ficarem sobrpostos e deixa o layout com tamanho flexivel
plt.tight_layout()
plt.show()
```

### Subplots - 2ª forma

```python
f, ax = plt.subplots(1, 2, figsize=(6,3), dpi=150) #1 linha, 2 colunas, tamanho da figura, qualidade da imagem gerada
ax[0].hist(data['Idade'].dropna())
ax[0].set_title('Distribuição das Idades')

ax[1].hist(data['Classe'].dropna())
ax[1].set_title('Distribuição das Classes')

# tira a distancia dos gráficos para não ficarem sobrpostos e deixa o layout com tamanho flexivel
plt.tight_layout()
plt.show()
```

## Seaborn

```python
import seaborn as sns
```

```python
##função value_counts() é do pandas
data['Sobreviveu'].value_counts()
```

```python
data['Idade'].value_counts()
```

## pie e countplot

```python
##o ax sempre será referenciado a quantidade de gráficos, neste exemplo serão dois
f, ax = plt.subplots(1, 2, figsize=(10,5))
data['Sobreviveu'].value_counts().plot.pie(ax=ax[0], explode=[0,0.05], autopct='%0.2f%%')
sns.countplot('Sexo', data=data, ax=ax[1])

plt.show()
```

```python
f, ax = plt.subplots(1, 2, figsize=(10,5 ))
data['Embarque'].value_counts().plot.pie(ax=ax[0], explode=[0.05, 0.05, 0.05], autopct='%0.2f%%')
sns.countplot('Embarque', data=data, ax=ax[1])

ax[0].set_title('% dos portões de embarque')
ax[1].set_title('Distribuição real por portão de embarque')
ax[1].set_ylabel('Total de passageiros')
ax[1].set_xlabel('Total por portão de embarque')

plt.tight_layout()
plt.show()
```

## countplot e hue

```python
sns.countplot('Sobreviveu', hue='Sexo', data=data)

plt.show()
```
