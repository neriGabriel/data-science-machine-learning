## População x Amostra

### População
Um grupo de pessoas ou objetos incluidos num estudo

#### Parametro
Caracteristica da populção, e.g: aultura minima, peso maximo, idade média e etc,
Notação estatistica: N -> população |   n -> amostra

### Variavel
Caracteristica de interesse coletado na amostra

### Dado
Valor do item em si

### Amostra
Pequena porção ou parte de populçao
Pode ser representativa ou tendenciosa

## Tipos de dados

### Categoricos
tipo de carro, cor e etc.

### Numérico
<b>Numeros discretos:</b> numeros de comodos, quantidade de filhos | geralmente são numeros inteiro, e.g: 1, 2, 3, 4 e etc.
<br><b>Numeros continuos:</b> numeros decimais, autlra, pesoa | geralmente são numeros decimais, e.g 10.5, 2.56 e etc.
<br>
### Qualitativo
<b>Nominal:</b> categorias, sem numeros, e.g: sim, não, rotulo, genêro e etc | geralmente não contém representativos categoricos.
<br><b>Ordinal:</b> ordem importa, porém não existe um espaço claro entre dois elementos, e.g: 1º lugar, 2º lugar e etc.
<br>
### Quantitativo
<b>Intervalar:</b> espaço entre dois são bem definidos e com siguinificado, e.g: temperatura de 10° graus é maior que 5° graus
<br><b>Razão:</b> zero é classificado como zero, e.g: pontuação em um teste, pesquisa sobre quantidade de filhos | o espaço é bem definido e etc.

## Técnicas de amostragens

### Regra #1
Amostra precisa ser representativa da população, aleatoria de preferencialpara não ser tendenciosa

### Métodos
<b>simples</b>,
<br><b>estratificada</b>,
<br><b>Por conglomerado</b>,
<br><b>Sistemática</b>,
<br><b>Por conveniência (não aleatorio)</b>,

#### Simples
Todos os individuos da população de uma amostra simples são escolhidos aleatoriamente.

#### Estratificada
Divide a popualçao em grupo e seleciona a proporção de cada na amostra.

#### Por conglomerado
Divide a popualçao em grupos e seleciona todos os individuos de um grupo aleatório, individuos de grupos não selecionado são excluidos da amostra.

#### Sistemática
Seleciona o primeiro individuos em ordem aleatorio e depois seleciona o enezimo numero. 

#### Por conveniência
Seleciona individuos que estão prontamente disponíves, e.g: entrevistas, amigos no local, lista de contatos, é considerado uma pesquisa tendenciosa.

## Freqência relativa

```python
def calcula_ticks(lista, barras):
    lista.sort()
    menor_valor = lista[0]
    maior_valor = lista[-1]
    intervalo = (maior_valor - menor_valor) / barras
    ticks = [menor_valor]
    ultimo = menor_valor+intervalo
    for x in range(barras-1):
        ultimo += intervalo
        ticks.append(ultimo)
    return ticks    
```

```python
km.plot.hist(rwidth=0.95)
ticks = calcula_ticks(km.values, 7)
plt.ticks = ticks
plt.title("Distancia de casa para tabalho")
plt.xlabel("Km")
plt.ylabel("Frenquencia")
plt.grid(axis='y')
plt.show()
```

```python
yticks = ax.get_yticks()
```

```python
## subistituir os labels do eixo y por %
def frequencia_relativa(total, yticks):
    freq_rel = []
    for item in yticks:
       freq_rel.append("{0:.2f}%".format( (item/total) * 100))
    return freq_rel
```

```python
lista = km.values
barras = 7
ax = km.plot.hist(bins=barras, rwidth=0.95)

total = len(lista)
freq_rel = frequencia_relativa(total, yticks)
ticks = calcula_ticks(lista, 7)
ax.set_yticklabels(freq_rel)
plt.xticks(ticks)
plt.title("Distância de casa para o trabalho")
plt.xlabel("KM")
plt.ylabel("Frequência")
plt.grid(axis='y')

plt.show()
```
