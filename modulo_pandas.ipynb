\s -> espaÃ§o em branco (tab, espaÃ§o, nova linha)

\b -> inicio de uma palavra
<br>
\B -> exceto inicio da palavra
<br>

$  -> final de uma string

<br>
|  -> ou
<br>
() -> grupo 
<br><br>
'*'     -> nenhum ou mais
<br>
'+'     -> um ou mais
<br>
?     -> nenhum ou um
<br>
{3}   -> numero exato
<br>
{3,4} -> range de numeros (minimo, maximo)

## Mudando valores linha a linha - iterrows()

```python
cabine = pd.read_csv('titanic/train.csv')
```

```python
cabine['Cabin'].head()
```

```python
for idx, _ in cabine[['Cabin']].dropna().iterrows():
    cabine['Cabin'].at[idx] = cabine['Cabin'].at[idx][0]
```

```python
cabine[['Cabin', 'Survived', 'Sex', 'Pclass']].dropna().groupby(['Cabin', 'Sex']).count()
```
