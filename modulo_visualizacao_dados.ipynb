## Preparando os dados

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('titanic/train.csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
data.columns = ['IdPassageiro', 'Sobreviveu', 'Classe', 'Nome', 'Sexo', 'Idade', 
                'IrmaosConjuge','PaisFilhos', 'Bilhete', 'Tarifa', 'Cabine', 'Embarque']
```

```python
data['Sexo'].replace({'male': 'homem', 'female':'mulher'}, inplace=True)
```

```python
data['Cabine'] = data['Cabine'].apply(lambda x: x[0] if pd.notna(x) else np.nan)
```

```python
data['Cabine']
```

## pie e countplot

```python
##o ax sempre será referenciado a quantidade de gráficos, neste exemplo serão dois
##o parametro explode sempre será um array referenciado ao espeçamento da quantidade de valores da informação
f, ax = plt.subplots(1, 2, figsize=(10,5))
data['Sobreviveu'].value_counts().plot.pie(ax=ax[0], explode=[0,0.05], autopct='%0.2f%%')
sns.countplot('Sexo', data=data, ax=ax[1])

plt.show()
```

```python
f, ax = plt.subplots(1, 2, figsize=(10,5 ))
data['Embarque'].value_counts().plot.pie(ax=ax[0], explode=[0.05, 0.05, 0.05], autopct='%0.2f%%')
sns.countplot('Embarque', data=data, ax=ax[1])

ax[0].set_title('% dos portões de embarque')
ax[1].set_title('Distribuição real por portão de embarque')
ax[1].set_ylabel('Total de passageiros')
ax[1].set_xlabel('Total por portão de embarque')

plt.tight_layout()
plt.show()
```

## FacetGrid com hist

```python
## A quantidade de colunas/graficos é dinamica, ou seja, conforme o parametro de informação passada é calculado automaticamente
## a quantidade de colunas
g = sns.FacetGrid(data, col='Sobreviveu')
g.map(plt.hist, 'Idade', bins=18)

plt.show()
```

```python
## o parametro alpha é equivalente a opacidade da coluna
## a linha acima das barras é o intervalo de confiança, por padrão é 95% e pode ser configurado pelo parametro ci=None
g = sns.FacetGrid(data, row='Embarque', col='Sobreviveu')
g.map(sns.barplot, 'Sexo', 'Tarifa', alpha=0.9)

##As configurações das figuras servem para a configuração do tamanho das barras/graficos
fig = plt.gcf()
fig.set_size_inches(10,6)
plt.show()
```

## FacetGrid com distplot

```python
## height=altura
## aspect=largura
g = sns.FacetGrid(data, col='Sobreviveu', row='Classe', height=1.8, aspect=2.2)
g.map(sns.distplot, 'Idade', bins=20)

plt.show()
```

## Catplot do tipo "point"

```python
##as barras são o intervalo de confiança
##os valores de confiança são baseados na quantidade de registros de cada coluna
##quanto maior a quantidade menos confiavel é
##o intervalo de confiança pode ser definido através do parametro ci=None
##por padrão é definido como 95%
sns.catplot('Embarque', 'Sobreviveu', data=data, kind='point')

##resoluação da imagem
fig=plt.gcf()
fig.set_size_inches(6,3)
plt.show()
```

```python
data['Embarque'].value_counts()
```

## FacetGrid com pointplot

```python
## as barras são o intervalo de confiança
##o intervalo de confiança pode ser definido através do parametro ci=None
##por padrão é definido como 95%
##pode ser definido pelo parametro ci=None
g = sns.FacetGrid(data, row='Embarque', height=1.8, aspect=2.2)
g.map(sns.pointplot, 'Classe', 'Sobreviveu', 'Sexo')
g.add_legend()
plt.show()
```

## Heatmap

```python
data.isnull().sum()
```

```python
##parametro yticklabels define se terá ou não label no eixo y
##parametro cbar define se terá ou não a barra de calor no eixo y a direita
##parametro cmap define a cor dos espaços vázios
sns.heatmap(data.isnull(), yticklabels=False, cbar=False, cmap='magma')

plt.show()
```

## Relplot

```python
## parametro x = eixo x
## parametro y = eixo y
## parametro hue = infomação
## parametro data = data
## parametro size = conforme a informação passada no size maior será o ponto
## parametro sizes = proporção do tamanho do ponto
## parametro alpha = opacidade
## parametro palette = cor
sns.relplot(x='Tarifa', y='Idade', hue='Sexo', data=data, size='Tarifa', sizes=(40,400), alpha=0.5, palette='magma')

plt.show()
```

## Jointplots tipo kde

```python
## kind='kde' serve para saber onde o dado está concentrado
## os gráficos da Laterais são outlayers
sns.jointplot('Idade', 'Tarifa', data=data, kind='kde', color='green')

plt.show()
```

```python
## kind='reg' serve para saber através de regressão linear
## os gráficos da Laterais são outlayers
sns.jointplot('Idade', 'Tarifa', data=data, kind='reg', color='green')

plt.show()
```

## Kdeplot

```python
## kde = curva de densidade
sns.kdeplot(data['Idade'].loc[data['Classe']==1], shade=True, color='g', label='1ª classe')
plt.show()
```

```python
## kde = curva de densidade
## aceita o parametro alpha
sns.kdeplot(data['Idade'].loc[data['Classe']==1], shade=True, color='g', label='1ª classe')
sns.kdeplot(data['Idade'].loc[data['Classe']==2], shade=True, color='r', label='2ª classe')
sns.kdeplot(data['Idade'].loc[data['Classe']==3], shade=True, color='b', label='3ª classe')

plt.figure(figsize=(8,5), dpi=80)
plt.show()
```

## Catplot, count e heatmap de correlações

```python
##col_wrap separa por linha a quantidade de graficos
sns.catplot('Sobreviveu', col='Cabine', data=data[['Sobreviveu', 'Cabine']].dropna(), kind='count', height=3.5, col_wrap=4
           , palette='tab20')


plt.show()
```

## Heatmap corr()

```python
##annot o quão relacionados esses dados estão
##mostra somente dados numéricos e não dados categoricos
plt.figure(figsize=(10,7))
sns.heatmap(data[['Classe', 'Idade', 'Sobreviveu', 'Tarifa', 'Embarque']].corr(), cmap="Blues", annot=True)

plt.show()
```

```python
data[['Classe', 'Idade', 'Sobreviveu', 'Tarifa', 'Embarque']].corr()
```

```python
## transformar dados categoricos 
data = pd.get_dummies(data=data, columns=['Embarque'], drop_first=True)
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(data[['Classe', 'Idade', 'Sobreviveu', 'Tarifa', 'Embarque_Q', 'Embarque_S']].corr(), cmap="Blues", annot=True)

plt.show()
```
